# Compare climate outcomes

Determine the difference between the non-altered scenarios that I am running and those produced emissions-driven using FaIRv1.6.2 and MAGICCv7.5.3 for IPCC AR6 WG3.

```python
import pandas as pd
import matplotlib.pyplot as pl
```

```python
df_ar6 = pd.read_csv('../data/wg3/ar6_snapshot_1690020777.csv')
```

```python
df_ar6
```

```python
df_ar6['Variable'][8]
```

```python
df_ar6.loc[
    (df_ar6['Scenario']=='SSP1-19')&
    (df_ar6['Variable']=='AR6 climate diagnostics|Surface Temperature (GSAT)|FaIRv1.6.2|50.0th Percentile'),
    '2000':
]
```

```python
df_ar6.columns[5:].astype(float)
```

```python
scenario = 'SSP1-19'
variable = 'AR6 climate diagnostics|Surface Temperature (GSAT)'
pl.fill_between(
    df_ar6.columns[5:].astype(float), 
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|FaIRv1.6.2|5.0th Percentile'),
        '2000':
    ].values.squeeze(),
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|FaIRv1.6.2|95.0th Percentile'),
        '2000':
    ].values.squeeze(),
    color='b',
    alpha=0.2,
    lw=0
)
pl.plot(
    df_ar6.columns[5:].astype(float), 
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|FaIRv1.6.2|50.0th Percentile'),
    '2000':
    ].values.squeeze(),
    color='b',
    label='FaIRv1.6.2'
)
pl.fill_between(
    df_ar6.columns[5:].astype(float), 
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|MAGICCv7.5.3|5.0th Percentile'),
        '2000':
    ].values.squeeze(),
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|MAGICCv7.5.3|95.0th Percentile'),
        '2000':
    ].values.squeeze(),
    color='r',
    alpha=0.2,
    lw=0
)
pl.plot(
    df_ar6.columns[5:].astype(float), 
    df_ar6.loc[
        (df_ar6['Scenario']==scenario)&
        (df_ar6['Variable']==f'{variable}|MAGICCv7.5.3|50.0th Percentile'),
    '2000':].values.squeeze(),
    color='r',
    label='MAGICCv7.5.3'
)
pl.title(scenario)
pl.legend()
```
